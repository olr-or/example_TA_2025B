```python
students = []        # [(이름, 학번, 전공)]
grades = []          # [(학번, 과목명, 점수)]
expenses = []        # [(학번, 날짜, 항목, 금액)]

def add_student():
    print("\n[학생 등록]")
    name = input("이름: ")
    sid = input("학번: ")
    major = input("전공: ")

    # 학번 중복 확인
    for s in students:
        if s[1] == sid:
            print("이미 존재하는 학번입니다.\n")
            return

    students.append((name, sid, major))
    print(f"{name} 학생 등록 완료.\n")


def list_students():
    print("\n[학생 목록]")
    if not students:
        print("등록된 학생 없음.\n")
        return
    for i, s in enumerate(students):
        print(f"{i+1}. 이름: {s[0]}, 학번: {s[1]}, 전공: {s[2]}")
    print()

def add_grade():
    print("\n[성적 입력]")
    sid = input("학번: ")

    if sid not in [s[1] for s in students]:
        print("존재하지 않는 학번입니다.\n")
        return

    subject = input("과목명: ")
    score = input("점수(0~100): ")

    if not score.isdigit() or not (0 <= int(score) <= 100):
        print("점수 입력 오류.\n")
        return

    grades.append((sid, subject, int(score)))
    print("성적 입력 완료.\n")


def analyze_grade():
    print("\n[성적 분석]")
    sid = input("학번: ")

    student_scores = [g for g in grades if g[0] == sid]

    if not student_scores:
        print("성적 데이터 없음.\n")
        return

    scores = [g[2] for g in student_scores]
    avg = sum(scores) / len(scores)

    max_sub = max(student_scores, key=lambda x: x[2])
    min_sub = min(student_scores, key=lambda x: x[2])

    if avg >= 90: grade = "A"
    elif avg >= 80: grade = "B"
    elif avg >= 70: grade = "C"
    elif avg >= 60: grade = "D"
    else: grade = "F"

    print(f"평균 점수: {avg:.2f} ({grade})")
    print(f"최고 점수 과목: {max_sub[1]} ({max_sub[2]})")
    print(f"최저 점수 과목: {min_sub[1]} ({min_sub[2]})\n")


def list_scores():
    print("\n[전체 성적 데이터]")
    if not grades:
        print("데이터 없음.\n")
        return
    for sid, subject, score in grades:
        print(f"학번 {sid}: {subject} = {score}")
    print()

def add_expense():
    print("\n[소비 기록 입력]")
    sid = input("학번: ")
    if sid not in [s[1] for s in students]:
        print("존재하지 않는 학번입니다.\n")
        return

    date = input("날짜(예: 2025-03-01): ")
    category = input("항목(식비/교통/도서/기타 등): ")
    cost = input("금액: ")

    if not cost.isdigit() or int(cost) <= 0:
        print("금액 입력 오류.\n")
        return

    expenses.append((sid, date, category, int(cost)))
    print("소비 데이터 저장 완료.\n")


def analyze_expense():
    print("\n[소비 분석]")
    sid = input("학번: ")

    student_exp = [e for e in expenses if e[0] == sid]

    if not student_exp:
        print("소비 데이터 없음.\n")
        return

    total = sum(e[3] for e in student_exp)
    print(f"총 소비 금액: {total}원")

    # 카테고리별 합계
    categories = {}
    for _, _, cat, cost in student_exp:
        if cat not in categories:
            categories[cat] = 0
        categories[cat] += cost

    print("\n카테고리별 소비 비중:")
    for cat, amt in categories.items():
        ratio = (amt / total) * 100
        print(f"- {cat}: {amt}원 ({ratio:.1f}%)")

    print()


def list_expenses():
    print("\n[전체 소비 데이터]")
    if not expenses:
        print("데이터 없음.\n")
        return
    for sid, date, cat, amt in expenses:
        print(f"{sid} | {date} | {cat} | {amt}원")
    print()

while True:
    print("========== 메인 메뉴 ==========")
    print("1. 학생 등록")
    print("2. 학생 목록 보기")
    print("3. 성적 입력")
    print("4. 성적 분석")
    print("5. 전체 성적 보기")
    print("6. 소비 기록 입력")
    print("7. 소비 분석")
    print("8. 전체 소비 보기")
    print("9. 종료")
    print("================================")

    choice = input("선택: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        list_students()
    elif choice == "3":
        add_grade()
    elif choice == "4":
        analyze_grade()
    elif choice == "5":
        list_scores()
    elif choice == "6":
        add_expense()
    elif choice == "7":
        analyze_expense()
    elif choice == "8":
        list_expenses()
    elif choice == "9":
        print("프로그램 종료.")
        break
    else:
        print("잘못된 입력.\n")
```
