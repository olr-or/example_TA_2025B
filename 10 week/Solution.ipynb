```python
# Solution 1
def replace_column_3x3(A, col, b):
    new = []
    i = 0
    while i < 3:
        row = []
        j = 0
        while j < 3:
            if j == col:
                row.append(b[i])
            else:
                row.append(A[i][j])
            j += 1
        new.append(row)
        i += 1
    return new

def solve_linear_3x3(A, b):
    detA = matrix_determinant(A)
    if detA == 0:
        return None   

    Ax = replace_column_3x3(A, 0, b)
    Ay = replace_column_3x3(A, 1, b)
    Az = replace_column_3x3(A, 2, b)

    dx = matrix_determinant(Ax)
    dy = matrix_determinant(Ay)
    dz = matrix_determinant(Az)

    return [dx/detA, dy/detA, dz/detA]

def test_Q1():
    print("\n================== [Q1 RESULT] ==================")
    A = [[2,1,3],[1,-1,4],[3,2,0]]
    b = [7,5,6]

    print_matrix("A",A)
    print("b =",b)

    x = solve_linear_3x3(A,b)
    print("\nSolution x =", x)

    Ax = apply_matrix_to_vector(A, x)
    print("A * x =", Ax)
    print("Verification:", Ax==b)
```

```python
# Solution 2
def abs_val(x):
    return x if x>=0 else -x

def check_orthogonal(v1,v2,v3):
    eps = 1e-6

    d12 = dot_product(v1,v2)
    d23 = dot_product(v2,v3)
    d31 = dot_product(v3,v1)

    ortho = (abs_val(d12)<=eps) and (abs_val(d23)<=eps) and (abs_val(d31)<=eps)

    M = [[v1[0],v1[1],v1[2]],
         [v2[0],v2[1],v2[2]],
         [v3[0],v3[1],v3[2]]]

    vol = abs_val(matrix_determinant(M))

    print("\n================== [Q2 RESULT] ==================")
    print_vector("v1",v1)
    print_vector("v2",v2)
    print_vector("v3",v3)
    print("\ndot(v1,v2) =",d12)
    print("dot(v2,v3) =",d23)
    print("dot(v3,v1) =",d31)
    print("\nVolume =",vol)
    print("Orthogonal =",ortho)
    return ortho, vol

def test_Q2():
    v1=[1,0,0]
    v2=[0,2,0]
    v3=[0,0,-3]
    check_orthogonal(v1,v2,v3)
```
